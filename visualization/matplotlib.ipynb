```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10 , 200) 
# 0,10---> data range parameter
# 200----> length of data
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y)
plt.xlabel("this is my X label")
plt.ylabel("this is my Y label")
plt.title("Title of Learn MPL")
plt.show()
```

### Another Example:

```python
a = np.random.rand(50)
```

```python
a
```

```python
b = np.random.rand(50)
```

```python
b
```

```python
plt.plot(a,b) 
```

##### This plot not working because of random data.
### So we use Scatter plot to find out data-points.

```python
plt.scatter(a,b)
```

```python
color = np.random.rand(50)
size = np.random.rand(50)
plt.scatter(a, b, c = color, s=size*400, alpha=.2) # alpha for opacity

plt.xlabel("This is my A label")
plt.ylabel("This is my B label")
plt.title("Random Data")
plt.show()
```
