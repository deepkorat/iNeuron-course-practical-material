```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10 , 200) 
# 0,10---> data range parameter
# 200----> length of data
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x, y)
```

#### Decoration tips

```python
plt.figure(figsize=(8,3)) # change full figure size
plt.plot(x,y, '--r') # try --b for blue color
plt.xlabel("this is my X label")
plt.ylabel("this is my Y label")
plt.title("Title of Learn MPL")

plt.grid() # to shwocase a background grid
plt.show()
```

# Scater Plot
### So we use Scatter plot to find out data-points.

```python
plt.scatter(a,b)
```

```python
color = np.random.rand(50)
size = np.random.rand(50)
plt.scatter(a, b, c = color, s=size*400, alpha=.2) # alpha for opacity

plt.xlabel("This is my A label")
plt.ylabel("This is my B label")
plt.title("Random Data")
plt.show()
```

#### Decoration tips

```python
# plt.scatter(a, b, c = 'r', s=size*400, alpha=.2) # red color
plt.scatter(a, b, c = '#2dd821', s=size*400, alpha=.2) # you can add color code also
```
